```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("dark_background")
```

```python
marks = np.random.randint(0,100,(5))
subs = ["Maths", "Physics", "Chemistry","Biology", "Computer"]
```

```python
plt.pie(marks, labels=subs)
```

```python
colors = ["blue", "orange","red","yellow","green"]
plt.pie(marks, labels=subs, colors=colors)
plt.show()
```

# Decimals
%0.1f%% - 1 Decimals \
%0.2f%% =  2 decimals

```python
plt.pie(marks, labels=subs, autopct="%0.1f%%")
plt.show()
```

# Explode

```python
explode = [0.1,0.2,0,0,0]
plt.pie(marks, labels=subs, autopct="%0.1f%%", explode=explode)
plt.show()
```

```python
plt.pie(marks, labels=subs, radius=1.6,  shadow=True)
plt.show()
```

# text props

```python
textprops = {'fontsize':16, "color":'b'}
plt.pie(marks, labels=subs, textprops=textprops,autopct="%0.1f%%")
plt.show()
```

# wedge props

```python
wedgeprops = {"linestyle":"--", "linewidth":2, "edgecolor":"white"}
plt.pie(marks, labels=subs,autopct="%0.1f%%",wedgeprops=wedgeprops)
plt.legend()
plt.show()
```
