```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread("tree.jpg")
plt.imshow(img)
plt.show()
```

# Open CV

```python
import cv2
```

```python
img = cv2.imread("tree.jpg")
plt.imshow(img)
plt.show()
```

As you can see the color is not accurate, opencv read in BGR format not RGB

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

# Aspect Ratio (h:w)

```python
plt.imshow(img, aspect=0.5)
```

# Color Bar
cmap defines color in color bar

```python
plt.imshow(img, aspect=0.5,cmap=("YlOrRd"))
plt.colorbar()
```
