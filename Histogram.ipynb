```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("dark_background")
```

```python
marks = np.random.randint(0,100,(50))
marks
```

```python
plt.hist(marks)
plt.show()
```

```python
bins = np.arange(0,100,5)
plt.hist(marks, bins=bins)
plt.show()
```

```python
plt.hist(marks, bins=bins, color="orange", orientation="horizontal")
plt.yticks(range(0,100,5))
plt.show()
```

```python
plt.hist(marks, rwidth=0.5)
```

```python
plt.hist(marks, rwidth=0.5, histtype="step")
```

```python
marks2 = np.random.randint(0,100,(50))
```

```python
plt.hist([marks,marks2], color=["blue","orange"])
```
