```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("dark_background")
```

```python
marks = [50,30, 40,20,15]
roll_no = [1,2,3,4,5]
plt.bar(roll_no, marks)
plt.show()
```

```python
colors = ["red","green", "blue", "yellow", "violet"]
plt.bar(roll_no, marks, color=colors, edgecolor="white", width=0.5)
plt.show()
```

```python
plt.bar(roll_no, marks, color=colors, edgecolor="white", linewidth=5, linestyle="--")
plt.show()
```

```python
plt.barh(roll_no, marks, color=colors)
```
