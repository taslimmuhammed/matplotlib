```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.style.use("dark_background")
```

```python
rollNo = range(1,10)
marks = range(10,100,10)
```

# Marker

```python
plt.scatter(rollNo, marks, color='green', marker="_") # makers = v / *
plt.show()
```

```python
plt.figure(figsize=(5,5))
sizes = range(10,100,10)
plt.scatter(rollNo, marks,marker='v', color='red', s=sizes, alpha=0.6)
plt.show()
```

# X and Y Axis

```python
plt.xticks(range(0,10))
plt.yticks(range(5,100,5))
plt.scatter(rollNo, marks)
plt.xlabel("Roll No")
plt.ylabel("Marks")
plt.show()
```
