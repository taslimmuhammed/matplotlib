```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("dark_background")
```

```python
marks = np.random.randint(0,100,(5))
subs = ["Phy", "Chem","Bio","Mat","Com"]
```

# subplot
plt.subplot(rows, columns, index)\
eg:- \
2*2 matrix\
1 | 2 \
3 | 4\
1,2,3,4, are the respective indexes


```python
plt.figure(figsize = (12,6))
plt.subplot(2,2,1)
plt.hist(marks, label=subs)
plt.subplot(2,2,2)
plt.plot(subs, marks)
plt.subplot(2,2,3)
plt.scatter(subs, marks)
plt.subplot(2,2,4)
plt.bar(subs,marks) 
```

# saving figure
dpi - dots per inch
transparent - transparent background

```python
plt.figure(figsize = (12,6))
plt.hist(marks, label=subs)
plt.savefig("graphs.png",dpi=300, transparent=True, bbox_inches='tight')
```
